### Apresentação dos dados

```python
import pandas as pd

df = pd.read_csv('notas_fiscais_exemplo.csv', sep=';')
display(df)
```

### Visualização dos produtos

```python
preco_de_cada_produto = df[['DescricaoProduto', 'ValorUnitario', 'Quantidade', 'ValorTotal']].groupby(['DescricaoProduto','ValorUnitario']).sum()
display(preco_de_cada_produto)
```

```python
products_description = df['DescricaoProduto'].unique().tolist()
results = {
    'DescricaoProduto': [], 'Média': [], 'Moda': [], 'Mediana': [], 
    'DesvioPadrao': []
    }

for product_description in products_description:
    results['DescricaoProduto'].append(product_description)
    produto_selecionado = df[df['DescricaoProduto'] == product_description]
    results['Média'].append(produto_selecionado['ValorUnitario'].mean())
    results['Moda'].append(produto_selecionado['ValorUnitario'].mode()[0])
    results['Mediana'].append(produto_selecionado['ValorUnitario'].median())
    results['DesvioPadrao'].append(produto_selecionado['ValorUnitario'].std())
pd.DataFrame(results)
```

```python
results_2 = {
    'DescricaoProduto': [], 'Média': [], 'Moda': [], 'Mediana': [], 
    'DesvioPadrao': []
    }

for product_description in products_description:
    results_2['DescricaoProduto'].append(product_description)
    produto_selecionado = df[df['DescricaoProduto'] == product_description]
    results_2['Média'].append(produto_selecionado['ValorTotal'].mean())
    results_2['Moda'].append(produto_selecionado['ValorTotal'].mode()[0])
    results_2['Mediana'].append(produto_selecionado['ValorTotal'].median())
    results_2['DesvioPadrao'].append(produto_selecionado['ValorTotal'].std())
pd.DataFrame(results_2)
```
