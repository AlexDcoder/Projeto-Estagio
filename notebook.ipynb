```python
import pandas as pd

df = pd.read_csv('notas_fiscais_exemplo.csv', sep=';')
display(df)
```

```python
products_description = df['DescricaoProduto'].unique().tolist()
results = {'DescricaoProduto': [], 'Média dos Preços': []}

for product_description in products_description:
    results['DescricaoProduto'].append(product_description)    
    produto_selecionado = df[df['DescricaoProduto'] == product_description]
    produto_selecionado[['DescricaoProduto', 'ValorUnitario', 'Quantidade', 'ValorTotal']].groupby(['DescricaoProduto','ValorUnitario']).sum()
    media_preco = (produto_selecionado['ValorTotal'].sum() / produto_selecionado['Quantidade'].sum())
    results['Média dos Preços'].append(media_preco)

pd.DataFrame(results)
```

```python
preco_de_cada_produto = df[['DescricaoProduto', 'ValorUnitario', 'Quantidade', 'ValorTotal']].groupby(['DescricaoProduto','ValorUnitario']).sum()
display(preco_de_cada_produto)
x = preco_de_cada_produto['ValorTotal'].sum()
y = preco_de_cada_produto['Quantidade'].sum()
print(f'Média dos preços dos produtos: R$ {(x / y):.2f}')
```
